# Scatter Plot
# Definition:
A scatter plot is a type of data visualization that uses dots to represent the values of two different numerical variables. Each dot on the plot corresponds to one observation in the dataset, with the position of the dot determined by the values of the two variables on the x and y axes.

# Purpose:
Scatter plots are used to observe and show relationships or correlations between the two variables. They can help in identifying patterns, trends, and potential outliers in the data.

# Example:
In a scatter plot showing the relationship between total_bill and tip in a restaurant dataset, each point would represent an individual transaction, with its position determined by the amount of the bill and the corresponding tip given.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('tips')
data
```

```python
# make scatter plot
sns.scatterplot(x='total_bill',y='tip',data=data)
plt.title("scatter plot")
plt.show()
```

# Line Plot
# Definition:
 A line plot (or line chart) is a type of data visualization that uses a continuous line to connect data points in a sequence. It shows the relationship between two variables, typically with the x-axis representing time or another continuous variable and the y-axis representing the measured value.

# Purpose:
 Line plots are commonly used to track changes over time or to show trends. They are ideal for visualizing the overall trend in a dataset, whether it's increasing, decreasing, or remaining stable.

# Example:
 In a line plot showing the relationship between total_bill and tip, the line would represent the general trend of how tips vary with the amount of the bill, helping to visualize the correlation between these two variables.
