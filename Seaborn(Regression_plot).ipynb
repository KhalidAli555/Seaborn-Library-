# Linear Regression
# Definition:
 An LM plot (short for linear model plot) is a type of plot that shows the relationship between two continuous variables along with a fitted regression line and a confidence interval. It combines elements of both scatter plots and line plots, where the scatter plot shows individual data points and the regression line shows the best fit linear relationship between the variables.

# Purpose:
LM plots are used to visualize and assess the relationship between two variables, specifically to see if there is a linear relationship. The plot also provides insights into the strength and direction of this relationship, as well as the confidence interval of the regression line.

# Example:
An LM plot could be used to visualize the relationship between the total bill and the tip amount in a restaurant dataset, showing the trend line that indicates how tips tend to increase as the bill amount increases.

# Load the dataset

```python
# import the library
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset
dataset = sns.load_dataset('tips')

# the first five entries of the dataset
dataset.head()
```

```python
sns.set_style('whitegrid') # for background color
sns.lmplot(x ='total_bill', y ='tip', data = dataset)
plt.show()
```

# give some extra feature or better understanding

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = dataset,
		hue ='sex', markers =['o', 'v'])
plt.show()
```
