# Displot
It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.
# Purpose
To display the distribution of a variable using a histogram, along with a KDE plot to estimate the probability density function.
# example
This plot shows the distribution of the total_bill with both a histogram and KDE plot overlay.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.set_style('whitegrid') # set the background style of the plot
sns.distplot(data['total_bill'], kde = True, color ='red', bins = 30) #KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn
plt.title("distplot")
plt.show()
```

# Joinplot
It is used to draw a plot of two variables with bivariate and univariate graphs. It basically combines two different plots.
# Purpose
To visualize the relationship between two variables along with their individual distributions.
It combines scatter plots, histograms, and kernel density estimates (KDE) in one plot.

```python
sns.jointplot(x ='total_bill', y ='tip',kind="scatter",data = data)
plt.title("Joinplot")
plt.show()
```

# pairplot
# Definition:
pairplot is a Seaborn function that creates a grid of scatter plots and histograms or KDEs for each pair of variables in a dataset.
# Purpose:
To explore relationships between multiple pairs of variables in a dataset.
Especially useful for identifying correlations or patterns among several variables.
# Example
 using tips dataset:

```python
sns.pairplot(data)
plt.show()
```

# rugplot
# Definition
rugplot is a Seaborn function that draws a small tick mark for each data point on a single axis, typically at the bottom of a distribution plot.

# Purpose
To provide a visual summary of the distribution of a dataset.
Often used to enhance other plots like KDEs or histograms by showing the exact positions of data points.

```python
 sns.rugplot(data['total_bill'])

plt.show()
```
